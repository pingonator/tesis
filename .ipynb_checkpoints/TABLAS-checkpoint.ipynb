```python
import pandas as pd
```

```python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s.upper()
```

```python
# fuente Anexo II RNE E.030

zona_sismica = pd.read_csv('zona_sismica.csv')

dict_zona_sismica = {}
for index, row in zona_sismica.iterrows():
    dict_zona_sismica[
        normalize(row['Departamento']), normalize(row['Provincia']), normalize(row['Distrito'])
    ] = row['Zona Sísmica']
zona_sismica
```
